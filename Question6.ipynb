second day, four pennies the third day, and so on with the daily pay doubling each day the employee works. The program should
ask the user for the number of days the employee worked during the month and should display a table showing how much the
salary was for each day worked, as well as the total pay earned for the month. The output should be displayed in dollars with
two decimal points, not in pennies.
Input Validation: Do not accept a number less than 1 or more than 31 for the number of days worked.

```python
salary=1
days=int(input("Enter the Number of Days you Worked in a month :"))
if days>1 and days<31 :
    for x in range(1,days+1):
        print(x,"       ",salary/100)
        salary=salary*2
```
